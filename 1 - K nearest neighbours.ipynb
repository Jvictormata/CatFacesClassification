```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
from sklearn.model_selection import train_test_split
import time
import torch
from sklearn.metrics import accuracy_score


from jax import grad,jit,random, jacfwd
import jax.numpy as jnp

%matplotlib inline
```

# Loading Data

```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            images.append(img.reshape(1,64*64))
    images = np.array(images)
    images = images.reshape(images.shape[0],64*64)/255
    return images
```

```python
X_cats = load_images_from_folder("dataset_cats/cats/")
y_cats = np.ones((X_cats.shape[0],1))
X_noncats = load_images_from_folder("dataset_cats/non-cats/")
y_noncats = np.zeros((X_noncats.shape[0],1))

X_data = np.append(X_cats,X_noncats, axis=0)
X_data = np.append(np.ones((X_data.shape[0],1)),X_data,axis = 1)
y_data = np.append(y_cats,y_noncats, axis=0)


X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=9)
X_train2, X_val, y_train2, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=9)
```

```python
# Example of a picture
index = 0
plt.imshow(X_train[index,1:].reshape((64,64)),cmap="gray")
print ("y = " + str(y_train[index]))
```

# K nearest neighbours - Validation

```python
def calc_class(element,X,y,k):
    dist = []
    for item in X:
        dist.append(np.linalg.norm(item-element))
        
    dist = np.array(dist)
    return y[dist.argsort()[:k]].mean()
```

```python
calc_class(X_test[0],X_train,y_train,10), y_test[0]
```

```python
predictions = []
Ks = np.array([1,5,10,15,20,25])

for i in Ks:
    pred = []
    for item in X_val:
        pred.append(calc_class(item,X_train2,y_train2,int(i)))
    pred = np.array(pred)
    predictions.append(pred)
```

```python
acc = []
for pred in predictions:
    acc.append(accuracy_score(y_val, pred>0.5))
acc = np.array(acc)
```

```python
plt.scatter(Ks,acc)
plt.xlabel("K",fontsize=16)
plt.ylabel("Validation accuracy",fontsize=16)
```
