```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
acc_KNN = np.load("acc_KNN.npy")
acc_logReg = np.load("acc_logReg.npy")
acc_NN = np.load("acc_NN.npy")

cost_logReg = np.load("cost_logReg.npy")
cost_NN = np.load("cost_NN.npy")
```

```python
ig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20,5))

ax[0].plot(np.linspace(acc_KNN,acc_KNN,21))
ax[0].plot(acc_logReg)
ax[0].plot(acc_NN)



ax[1].plot(cost_logReg)
ax[1].plot(cost_NN)


ax[0].set_xlabel("Epoch",fontsize=16)
ax[0].set_ylabel("Test accuracy",fontsize=16)
ax[1].set_ylabel("Training loss",fontsize=16)
ax[1].set_xlabel("Iteration",fontsize=16)

ax[0].tick_params(axis="x", labelsize=14) 
ax[0].tick_params(axis="y", labelsize=14) 
ax[1].tick_params(axis="x", labelsize=14) 
ax[1].tick_params(axis="y", labelsize=14) 
```

```python
plt.figure(figsize=(16,6))

plt.plot(np.linspace(acc_KNN,acc_KNN,21))
plt.plot(acc_logReg)
plt.plot(acc_NN)

plt.xlabel("Epoch",fontsize=16)
plt.ylabel("Test accuracy",fontsize=16)

legend = ["KNN",
          "Log Reg",
          "Neural Net"
             ]

plt.legend(legend,fontsize=16)

plt.savefig('result.pdf')  
```
